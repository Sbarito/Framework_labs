```python
def preprocess_student_classification(df):
    """ะัะตะดะพะฑัะฐะฑะพัะบะฐ ัััะดะตะฝัะตัะบะธั ะดะฐะฝะฝัั ะดะปั ะบะปะฐััะธัะธะบะฐัะธะธ"""
    print("\n" + "="*50)
    print("ะะะะะะะะะะะขะะ ะกะขะฃะะะะงะะกะะะฅ ะะะะะซะฅ (ะะะะกะกะะคะะะะฆะะฏ)")
    print("="*50)
    
    if df is None:
        print("โ DataFrame ะฟัััะพะน")
        return None, None, None
    
    df_clean = df.copy()
    
    print("ะะพัััะฟะฝัะต ะบะพะปะพะฝะบะธ:")
    for i, col in enumerate(df_clean.columns):
        print(f"  {i+1:2d}. '{col}' (ัะธะฟ: {df_clean[col].dtype})")
    
    if 'G3' not in df_clean.columns:
        print("โ ะะพะปะพะฝะบะฐ 'G3' ะฝะต ะฝะฐะนะดะตะฝะฐ!")
        print("   ะะพัััะฟะฝัะต ะบะพะปะพะฝะบะธ:", list(df_clean.columns))
        return None, None, None
    
    print(f"โ ะะฐะนะดะตะฝะฐ ัะตะปะตะฒะฐั ะฟะตัะตะผะตะฝะฝะฐั: 'G3'")
    print(f"  ะะธะฐะฟะฐะทะพะฝ ะทะฝะฐัะตะฝะธะน: {df_clean['G3'].min()} - {df_clean['G3'].max()}")
    
    df_clean['academic_success'] = (df_clean['G3'] >= 10).astype(int)
    success_rate = df_clean['academic_success'].mean() * 100
    print(f"โ ะกะพะทะดะฐะฝะฐ ัะตะปะตะฒะฐั ะฟะตัะตะผะตะฝะฝะฐั 'academic_success'")
    print(f"  ะัะพัะตะฝั ััะฟะตะฒะฐััะธั ัััะดะตะฝัะพะฒ: {success_rate:.1f}%")
    
    numeric_features = []
    possible_numeric = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 
                       'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']
    
    print("\nะะพะธัะบ ัะธัะปะพะฒัั ะฟัะธะทะฝะฐะบะพะฒ:")
    for feature in possible_numeric:
        if feature in df_clean.columns:
            if df_clean[feature].dtype == 'object':
                df_clean[feature] = pd.to_numeric(df_clean[feature], errors='coerce')
            numeric_features.append(feature)
            print(f"  โ '{feature}' ะฝะฐะนะดะตะฝ")
        else:
            print(f"  โ '{feature}' ะฝะต ะฝะฐะนะดะตะฝ")
    
    categorical_features = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 
                           'Fjob', 'reason', 'guardian']
    
    features = numeric_features.copy()
    le = LabelEncoder()
    
    print("\nะะพะธัะบ ะธ ะบะพะดะธัะพะฒะฐะฝะธะต ะบะฐัะตะณะพัะธะฐะปัะฝัั ะฟัะธะทะฝะฐะบะพะฒ:")
    for feature in categorical_features:
        if feature in df_clean.columns:
            try:
                df_clean[f'{feature}_encoded'] = le.fit_transform(df_clean[feature].astype(str))
                features.append(f'{feature}_encoded')
                print(f"  โ '{feature}' ะทะฐะบะพะดะธัะพะฒะฐะฝ")
            except Exception as e:
                print(f"  โ '{feature}' ะพัะธะฑะบะฐ ะบะพะดะธัะพะฒะฐะฝะธั: {e}")
        else:
            print(f"  โ '{feature}' ะฝะต ะฝะฐะนะดะตะฝ")
    
    initial_count = len(df_clean)
    df_clean = df_clean.dropna(subset=features + ['academic_success'])
    final_count = len(df_clean)
    
    print(f"\nโ ะัะธััะบะฐ ะดะฐะฝะฝัั: {initial_count} โ {final_count} ัััะพะบ")
    print(f"โ ะัะฟะพะปัะทัะตััั {len(features)} ะฟัะธะทะฝะฐะบะพะฒ")
    
    if not features or df_clean.empty:
        print("โ ะะตะดะพััะฐัะพัะฝะพ ะดะฐะฝะฝัั ะดะปั ะฐะฝะฐะปะธะทะฐ")
        return None, None, None
    
    X = df_clean[features]
    y = df_clean['academic_success']
    
    print(f"โ ะัะพะณะพะฒัะน ะฝะฐะฑะพั ะดะฐะฝะฝัั: {X.shape[0]} samples, {X.shape[1]} features")
    print(f"  ะะฐะปะฐะฝั ะบะปะฐััะพะฒ: {np.bincount(y)}")
    
    return X, y, features
```

ะคัะฝะบัะธะธ ะฟัะตะดะพะฑัะฐะฑะพัะบะธ ะดะฐะฝะฝัั ะดะปั ัะตะณัะตััะธะธ (ะดะฐะฝะฝัะต ะพ ะฝะตะดะฒะธะถะธะผะพััะธ)

```python
def preprocess_house_regression(df):
    """ะัะตะดะพะฑัะฐะฑะพัะบะฐ ะดะฐะฝะฝัั ะพ ะฝะตะดะฒะธะถะธะผะพััะธ ะดะปั ัะตะณัะตััะธะธ"""
    print("\n" + "="*50)
    print("ะะะะะะะะะะะขะะ ะะะะะซะฅ ะ ะะะะะะะะะะกะขะ (ะะะะะะกะกะะฏ)")
    print("="*50)
    
    if df is None:
        return None, None, None
    
    df_clean = df.copy()
    
    print(f"โ ะฆะตะปะตะฒะฐั ะฟะตัะตะผะตะฝะฝะฐั: Rent (ะฐัะตะฝะดะฝะฐั ะฟะปะฐัะฐ)")
    print(f"  ะะธะฐะฟะฐะทะพะฝ ัะตะฝ: {df_clean['Rent'].min()} - {df_clean['Rent'].max()}")
    
    if 'Floor' in df_clean.columns:
        def parse_floor(floor_str):
            if pd.isna(floor_str):
                return 0
            floor_str = str(floor_str)
            if 'Ground' in floor_str:
                return 0
            if 'Basement' in floor_str:
                return -1
            try:
                parts = floor_str.split()
                for part in parts:
                    if part.isdigit():
                        return int(part)
                return 0
            except:
                return 0
        
        df_clean['floor_number'] = df_clean['Floor'].apply(parse_floor)
        print("โ ะะฑัะฐะฑะพัะฐะฝะฐ ะบะพะปะพะฝะบะฐ Floor")
    
    if 'Size' in df_clean.columns:
        try:
            df_clean['Size'] = df_clean['Size'].astype(str).str.replace(' sqft', '').astype(float)
            print("โ ะะฑัะฐะฑะพัะฐะฝะฐ ะบะพะปะพะฝะบะฐ Size")
        except Exception as e:
            print(f"โ ะัะธะฑะบะฐ ะพะฑัะฐะฑะพัะบะธ Size: {e}")
    
    features = ['BHK', 'Bathroom']
    if 'Size' in df_clean.columns:
        features.append('Size')
    if 'floor_number' in df_clean.columns:
        features.append('floor_number')
    

    categorical_cols = ['Area Type', 'City', 'Furnishing Status', 'Tenant Preferred', 'Point of Contact']
    le = LabelEncoder()
    
    for col in categorical_cols:
        if col in df_clean.columns:
            try:
                df_clean[col + '_encoded'] = le.fit_transform(df_clean[col].astype(str))
                features.append(col + '_encoded')
                print(f"โ ะะฐะบะพะดะธัะพะฒะฐะฝะฐ ะบะพะปะพะฝะบะฐ: {col}")
            except Exception as e:
                print(f"โ ะัะธะฑะบะฐ ะบะพะดะธัะพะฒะฐะฝะธั {col}: {e}")
    
    X = df_clean[features]
    y = df_clean['Rent']
    
    print(f"โ ะัะพะณะพะฒัะน ะฝะฐะฑะพั ะดะฐะฝะฝัั: {X.shape[0]} samples, {X.shape[1]} features")
    print(f"  ะกัะฐัะธััะธะบะฐ ะฐัะตะฝะดะฝะพะน ะฟะปะฐัั: mean={y.mean():.2f}, std={y.std():.2f}")
    
    return X, y, features
```

ะคัะฝะบัะธั ะธััะปะตะดะพะฒะฐะฝะธั KNN ะดะปั ะบะปะฐััะธัะธะบะฐัะธะธ

```python
def knn_classification_study(X, y, dataset_name="ะะฐะฝะฝัะต"):
    """ะััะปะตะดะพะฒะฐะฝะธะต KNN ะดะปั ะบะปะฐััะธัะธะบะฐัะธะธ"""
    print(f"\n" + "="*70)
    print(f"KNN ะะะะกะกะะคะะะะฆะะฏ - {dataset_name}")
    print("="*70)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    print("๐ ะััะปะตะดะพะฒะฐะฝะธะต ะฒะปะธัะฝะธั ะฟะฐัะฐะผะตััะฐ k ะฝะฐ ัะพัะฝะพััั...")
    k_range = range(1, 21)
    train_scores = []
    test_scores = []
    
    for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train_scaled, y_train)
        train_scores.append(knn.score(X_train_scaled, y_train))
        test_scores.append(knn.score(X_test_scaled, y_test))
    
    best_k_index = np.argmax(test_scores)
    best_k = k_range[best_k_index]
    best_score = test_scores[best_k_index]
    best_train_score = train_scores[best_k_index]
    
    print(f"๐ฏ ะะตะทัะปััะฐัั:")
    print(f"  ะะฟัะธะผะฐะปัะฝะพะต k: {best_k}")
    print(f"  ะัััะฐั ัะพัะฝะพััั: {best_score:.4f}")
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(k_range, train_scores, label='ะะฑััะฐััะฐั ะฒัะฑะพัะบะฐ', marker='o', linewidth=2)
    plt.plot(k_range, test_scores, label='ะขะตััะพะฒะฐั ะฒัะฑะพัะบะฐ', marker='s', linewidth=2)
    plt.axvline(x=best_k, color='red', linestyle='--', alpha=0.7, label=f'ะัััะตะต k={best_k}')
    plt.xlabel('ะะพะปะธัะตััะฒะพ ัะพัะตะดะตะน (k)')
    plt.ylabel('ะขะพัะฝะพััั')
    plt.title(f'KNN ะะปะฐััะธัะธะบะฐัะธั - ะะปะธัะฝะธะต ะฟะฐัะฐะผะตััะฐ k\n{dataset_name}', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    best_knn = KNeighborsClassifier(n_neighbors=best_k)
    best_knn.fit(X_train_scaled, y_train)
    y_pred = best_knn.predict(X_test_scaled)
    
    plt.subplot(1, 2, 2)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['ะะตััะฟะตั', 'ะฃัะฟะตั'], 
                yticklabels=['ะะตััะฟะตั', 'ะฃัะฟะตั'])
    plt.title(f'ะะฐััะธัะฐ ะพัะธะฑะพะบ (k={best_k})\nะขะพัะฝะพััั: {best_score:.4f}', fontsize=12)
    plt.ylabel('ะคะฐะบัะธัะตัะบะธะต ะทะฝะฐัะตะฝะธั')
    plt.xlabel('ะัะตะดัะบะฐะทะฐะฝะฝัะต ะทะฝะฐัะตะฝะธั')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n๐ ะะะะะะ KNN ะะะะะะ:")
    print(f"  ะะฟัะธะผะฐะปัะฝะพะต ะบะพะปะธัะตััะฒะพ ัะพัะตะดะตะน: {best_k}")
    print(f"  ะขะพัะฝะพััั ะฝะฐ ัะตััะพะฒะพะน ะฒัะฑะพัะบะต: {best_score:.4f}")
    print(f"  ะขะพัะฝะพััั ะฝะฐ ะพะฑััะฐััะตะน ะฒัะฑะพัะบะต: {best_train_score:.4f}")
    
    overfitting_gap = best_train_score - best_score
    print(f"  ะะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test accuracy: {overfitting_gap:.4f}")
    
    if overfitting_gap > 0.1:
        print("  โ๏ธ ะะพะทะผะพะถะฝะพะต ะฟะตัะตะพะฑััะตะฝะธะต (ะฑะพะปััะฐั ัะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test)")
    else:
        print("  โ ะะพะดะตะปั ัะพัะพัะพ ะพะฑะพะฑัะฐะตั (ะผะฐะปะฐั ัะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test)")
    
    from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
    
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    print(f"\n๐ ะะะะะะะะขะะะฌะะซะ ะะะขะะะะ ะะะะกะกะะคะะะะฆะะ:")
    print(f"  Precision (ะขะพัะฝะพััั): {precision:.4f}")
    print(f"  Recall (ะะพะปะฝะพัะฐ): {recall:.4f}")
    print(f"  F1-Score: {f1:.4f}")
    
    if best_score >= 0.9:
        print("  ๐ ะัะปะธัะฝะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (ะขะพัะฝะพััั โฅ 0.9)")
    elif best_score >= 0.8:
        print("  ๐ ะฅะพัะพัะตะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.8 โค ะขะพัะฝะพััั < 0.9)")
    elif best_score >= 0.7:
        print("  ๐ ะฃะดะพะฒะปะตัะฒะพัะธัะตะปัะฝะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.7 โค ะขะพัะฝะพััั < 0.8)")
    else:
        print("  ๐ก ะะธะทะบะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (ะขะพัะฝะพััั < 0.7)")
    
    print(f"\n๐ ะะะขะะะฌะะซะ ะะขะงะะข ะะ ะะะะกะกะะ:")
    print(classification_report(y_test, y_pred, target_names=['ะะตััะฟะตั', 'ะฃัะฟะตั']))
    
    return best_knn, best_k, best_score
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
def knn_regression_study(X, y, dataset_name="ะะฐะฝะฝัะต"):
    """ะััะปะตะดะพะฒะฐะฝะธะต KNN ะดะปั ัะตะณัะตััะธะธ"""
    print(f"\n" + "="*70)
    print(f"KNN ะะะะะะกะกะะฏ - {dataset_name}")
    print("="*70)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    print("๐ ะััะปะตะดะพะฒะฐะฝะธะต ะฒะปะธัะฝะธั ะฟะฐัะฐะผะตััะฐ k ะฝะฐ Rยฒ score...")
    k_range = range(1, 21)
    train_scores = []
    test_scores = []
    
    for k in k_range:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(X_train_scaled, y_train)
        train_scores.append(knn.score(X_train_scaled, y_train))
        test_scores.append(knn.score(X_test_scaled, y_test))
    
    best_k_index = np.argmax(test_scores)
    best_k = k_range[best_k_index]
    best_score = test_scores[best_k_index]
    best_train_score = train_scores[best_k_index]
    
    print(f"๐ฏ ะะตะทัะปััะฐัั:")
    print(f"  ะะฟัะธะผะฐะปัะฝะพะต k: {best_k}")
    print(f"  ะัััะธะน Rยฒ score: {best_score:.4f}")
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(k_range, train_scores, label='ะะฑััะฐััะฐั ะฒัะฑะพัะบะฐ', marker='o', linewidth=2)
    plt.plot(k_range, test_scores, label='ะขะตััะพะฒะฐั ะฒัะฑะพัะบะฐ', marker='s', linewidth=2)
    plt.axvline(x=best_k, color='red', linestyle='--', alpha=0.7, label=f'ะัััะตะต k={best_k}')
    plt.xlabel('ะะพะปะธัะตััะฒะพ ัะพัะตะดะตะน (k)')
    plt.ylabel('Rยฒ Score')
    plt.title(f'KNN ะะตะณัะตััะธั - ะะปะธัะฝะธะต ะฟะฐัะฐะผะตััะฐ k\n{dataset_name}', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    best_knn = KNeighborsRegressor(n_neighbors=best_k)
    best_knn.fit(X_train_scaled, y_train)
    y_pred = best_knn.predict(X_test_scaled)
    
    plt.subplot(1, 2, 2)
    plt.scatter(y_test, y_pred, alpha=0.6, s=50)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
    plt.xlabel('ะคะฐะบัะธัะตัะบะธะต ะทะฝะฐัะตะฝะธั')
    plt.ylabel('ะัะตะดัะบะฐะทะฐะฝะฝัะต ะทะฝะฐัะตะฝะธั')
    plt.title(f'ะัะตะดัะบะฐะทะฐะฝะธั vs ะคะฐะบัะธัะตัะบะธะต ะทะฝะฐัะตะฝะธั\nRยฒ = {best_score:.4f}', fontsize=12)
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n๐ ะะะะะะ KNN ะะะะะะ:")
    print(f"  ะะฟัะธะผะฐะปัะฝะพะต ะบะพะปะธัะตััะฒะพ ัะพัะตะดะตะน: {best_k}")
    print(f"  Rยฒ score ะฝะฐ ัะตััะพะฒะพะน ะฒัะฑะพัะบะต: {best_score:.4f}")
    print(f"  Rยฒ score ะฝะฐ ะพะฑััะฐััะตะน ะฒัะฑะพัะบะต: {best_train_score:.4f}")
    
    overfitting_gap = best_train_score - best_score
    print(f"  ะะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test Rยฒ: {abs(overfitting_gap):.4f}")
    
    if overfitting_gap > 0.1:
        print("  โ๏ธ ะะพะทะผะพะถะฝะพะต ะฟะตัะตะพะฑััะตะฝะธะต (ะฑะพะปััะฐั ัะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test)")
    else:
        print("  โ ะะพะดะตะปั ัะพัะพัะพ ะพะฑะพะฑัะฐะตั (ะผะฐะปะฐั ัะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test)")
    
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    
    print(f"\n๐ ะะะะะะะะขะะะฌะะซะ ะะะขะะะะ ะะะะะะกะกะะ:")
    print(f"  MSE (Mean Squared Error): {mse:.2f}")
    print(f"  RMSE (Root Mean Squared Error): {rmse:.2f}")
    print(f"  MAE (Mean Absolute Error): {mae:.2f}")
    
    if best_score >= 0.7:
        print("  ๐ ะัะปะธัะฝะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (Rยฒ โฅ 0.7)")
    elif best_score >= 0.5:
        print("  ๐ ะฅะพัะพัะตะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.5 โค Rยฒ < 0.7)")
    elif best_score >= 0.3:
        print("  ๐ ะฃะดะพะฒะปะตัะฒะพัะธัะตะปัะฝะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.3 โค Rยฒ < 0.5)")
    else:
        print("  ๐ก ะะธะทะบะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (Rยฒ < 0.3)")
    
    return best_knn, best_k, best_score
```

```python
def advanced_preprocessing(df):
    """ะฃะปัััะตะฝะฝัะน ะฟัะตะฟัะพัะตััะธะฝะณ ั ะพะฑัะฐะฑะพัะบะพะน ะฒัะฑัะพัะพะฒ ะผะตัะพะดะพะผ IQR"""
    numeric_columns = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 
                      'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences',
                      'G1', 'G2', 'G3']
    
    numeric_columns = [col for col in numeric_columns if col in df.columns]
    
    print(f"๐ง ะะะะะะะขะะ ะะซะะะะกะะ ะ {len(numeric_columns)} ะงะะกะะะะซะฅ ะะะะะะะะฅ...")
    
    initial_size = len(df)
    print(f"  ะะฐะทะผะตั ะดะฐะฝะฝัั ะดะพ ะพะฑัะฐะฑะพัะบะธ: {initial_size} ัััะพะบ")
    
    Q1 = df[numeric_columns].quantile(0.25)
    Q3 = df[numeric_columns].quantile(0.75)
    IQR = Q3 - Q1
    
    outlier_mask = ((df[numeric_columns] < (Q1 - 1.5 * IQR)) | 
                    (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)
    
    print(f"  ะะฐะนะดะตะฝะพ ัััะพะบ ั ะฒัะฑัะพัะฐะผะธ: {outlier_mask.sum()}")
    print(f"  ะัะพัะตะฝั ะฒัะฑัะพัะพะฒ: {outlier_mask.sum()/initial_size*100:.1f}%")
    
    df_clean = df[~outlier_mask].copy()
    
    print("  ๐ ะะฐะฟะพะปะฝะตะฝะธะต ะฟัะพะฟััะบะพะฒ ะฒ ัะธัะปะพะฒัั ะบะพะปะพะฝะบะฐั...")
    for col in numeric_columns:
        if df_clean[col].isna().any():
            median_val = df_clean[col].median()
            df_clean[col] = df_clean[col].fillna(median_val)
            print(f"    โ ะะฐะฟะพะปะฝะตะฝั ะฟัะพะฟััะบะธ ะฒ '{col}' ะผะตะดะธะฐะฝะพะน: {median_val}")
    
    print(f"  โ ะะฐะทะผะตั ะดะฐะฝะฝัั ะฟะพัะปะต ะพะฑัะฐะฑะพัะบะธ: {len(df_clean)} ัััะพะบ")
    print(f"  โ ะฃะดะฐะปะตะฝะพ ัััะพะบ: {initial_size - len(df_clean)}")
    print(f"  โ ะกะพััะฐะฝะตะฝะพ ะดะฐะฝะฝัั: {len(df_clean)/initial_size*100:.1f}%")
    
    return df_clean

def advanced_preprocessing_house(df):
    """ะฃะปัััะตะฝะฝัะน ะฟัะตะฟัะพัะตััะธะฝะณ ั ะพะฑัะฐะฑะพัะบะพะน ะฒัะฑัะพัะพะฒ ะผะตัะพะดะพะผ IQR ะดะปั house ะดะฐะฝะฝัั"""
    numeric_columns = ['BHK', 'Bathroom', 'Size', 'Rent']
    
    numeric_columns = [col for col in numeric_columns if col in df.columns]
    
    print(f"๐ง ะะะะะะะขะะ ะะซะะะะกะะ ะ {len(numeric_columns)} ะงะะกะะะะซะฅ ะะะะะะะะฅ...")
    
    initial_size = len(df)
    print(f"  ะะฐะทะผะตั ะดะฐะฝะฝัั ะดะพ ะพะฑัะฐะฑะพัะบะธ: {initial_size} ัััะพะบ")
    
    Q1 = df[numeric_columns].quantile(0.25)
    Q3 = df[numeric_columns].quantile(0.75)
    IQR = Q3 - Q1
    
    outlier_mask = ((df[numeric_columns] < (Q1 - 1.5 * IQR)) | 
                    (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)
    
    print(f"  ะะฐะนะดะตะฝะพ ัััะพะบ ั ะฒัะฑัะพัะฐะผะธ: {outlier_mask.sum()}")
    print(f"  ะัะพัะตะฝั ะฒัะฑัะพัะพะฒ: {outlier_mask.sum()/initial_size*100:.1f}%")
    
    df_clean = df[~outlier_mask].copy()
    
    print("  ๐ ะะฐะฟะพะปะฝะตะฝะธะต ะฟัะพะฟััะบะพะฒ ะฒ ัะธัะปะพะฒัั ะบะพะปะพะฝะบะฐั...")
    for col in numeric_columns:
        if df_clean[col].isna().any():
            median_val = df_clean[col].median()
            df_clean[col] = df_clean[col].fillna(median_val)
            print(f"    โ ะะฐะฟะพะปะฝะตะฝั ะฟัะพะฟััะบะธ ะฒ '{col}' ะผะตะดะธะฐะฝะพะน: {median_val}")
    
    print(f"  โ ะะฐะทะผะตั ะดะฐะฝะฝัั ะฟะพัะปะต ะพะฑัะฐะฑะพัะบะธ: {len(df_clean)} ัััะพะบ")
    print(f"  โ ะฃะดะฐะปะตะฝะพ ัััะพะบ: {initial_size - len(df_clean)}")
    print(f"  โ ะกะพััะฐะฝะตะฝะพ ะดะฐะฝะฝัั: {len(df_clean)/initial_size*100:.1f}%")
    
    return df_clean
```

ะธะฝะถะตะฝะตัั ะฟัะธะทะฝะฐะบะพะฒ (ัะพะทะดะฐะฝะธะต ะฝะพะฒัั ะฟัะธะทะฝะฐะบะพะฒ ะฟะพะฒััะธั ะฟัะตะดัะบะฐะทะฐัะตะปัะฝัั ัะธะปั)

```python
def feature_engineering(df):
    """ะกะพะทะดะฐะฝะธะต ะฝะพะฒัั ะฟัะธะทะฝะฐะบะพะฒ ะดะปั ัะปัััะตะฝะธั ะผะพะดะตะปะตะน"""
    print("๐ง ะกะะะะะะะ ะะะะซะฅ ะะะะะะะะะ...")
    
    df_fe = df.copy()
    
    if 'G1' in df.columns and 'G2' in df.columns:
        df_fe['grade_trend'] = df_fe['G2'] - df_fe['G1']
        print("  โ ะกะพะทะดะฐะฝ ะฟัะธะทะฝะฐะบ 'grade_trend' (ััะตะฝะด ััะฟะตะฒะฐะตะผะพััะธ)")
    
    if 'Dalc' in df.columns and 'Walc' in df.columns:
        df_fe['total_alcohol'] = df_fe['Dalc'] + df_fe['Walc']
        print("  โ ะกะพะทะดะฐะฝ ะฟัะธะทะฝะฐะบ 'total_alcohol' (ะพะฑัะตะต ะฟะพััะตะฑะปะตะฝะธะต ะฐะปะบะพะณะพะปั)")
    
    if 'studytime' in df.columns and 'goout' in df.columns:
        df_fe['study_effort'] = df_fe['studytime'] * (1 - df_fe['goout']/5)
        print("  โ ะกะพะทะดะฐะฝ ะฟัะธะทะฝะฐะบ 'study_effort' (ะธะฝะดะตะบั ััะตะฑะฝัั ััะธะปะธะน)")
    
    if 'Size' in df.columns and 'BHK' in df.columns and 'Rent' in df.columns:
        df_fe['price_per_sqft'] = df_fe['Rent'] / df_fe['Size']
        print("  โ ะกะพะทะดะฐะฝ ะฟัะธะทะฝะฐะบ 'price_per_sqft' (ัะตะฝะฐ ะทะฐ ะบะฒ.ััั)")
        
        df_fe['room_density'] = df_fe['BHK'] / df_fe['Size']
        print("  โ ะกะพะทะดะฐะฝ ะฟัะธะทะฝะฐะบ 'room_density' (ะฟะปะพัะฝะพััั ะบะพะผะฝะฐั)")
    
    return df_fe
```

ะฒะธะทัะฐะปะธะทะฐัะธั ะธ EDA (ะณะปัะฑะพะบะธะน ะฐะฝะฐะปะธะท ะดะฐะฝะฝัั ะฒััะฒะธั ัะบััััะต ะทะฐะฒะธัะธะผะพััะธ)

```python
def evaluate_with_eda_insights(X, y, problem_type='classification', dataset_name="ะะฐะฝะฝัะต"):
    """ะัะตะฝะบะฐ ะผะพะดะตะปะธ ั ะธัะฟะพะปัะทะพะฒะฐะฝะธะตะผ ะธะฝัะฐะนัะพะฒ ะธะท EDA"""
    print(f"\n๐ ะะกะะะะฌะะะะะะะ ะะะกะะะขะะ ะะ EDA ะะะฏ {dataset_name}")
    
    if problem_type == 'classification':
        from sklearn.ensemble import RandomForestClassifier
        model = RandomForestClassifier(n_estimators=100, random_state=42)
    else:
        from sklearn.ensemble import RandomForestRegressor
        model = RandomForestRegressor(n_estimators=100, random_state=42)
    
    model.fit(X, y)
    feature_imp = pd.Series(model.feature_importances_, index=X.columns)
    
    top_features = feature_imp.nlargest(int(len(feature_imp) * 0.8)) 
    X_reduced = X[top_features.index]
    
    print(f"  โ ะัะพะฑัะฐะฝะพ {len(top_features)} ัะฐะผัั ะฒะฐะถะฝัั ะฟัะธะทะฝะฐะบะพะฒ ะธะท {len(feature_imp)}")
    print(f"  โ ะขะพะฟ-3 ะฟัะธะทะฝะฐะบะฐ: {list(top_features.head(3).index)}")
    
    if problem_type == 'classification':
        X_train, X_test, y_train, y_test = train_test_split(
            X_reduced, y, test_size=0.3, random_state=42, stratify=y
        )
        
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        knn = KNeighborsClassifier(n_neighbors=5)
        knn.fit(X_train_scaled, y_train)
        score = knn.score(X_test_scaled, y_test)
        
    else:
        X_train, X_test, y_train, y_test = train_test_split(
            X_reduced, y, test_size=0.3, random_state=42
        )
        
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        knn = KNeighborsRegressor(n_neighbors=5)
        knn.fit(X_train_scaled, y_train)
        score = knn.score(X_test_scaled, y_test)
    
    print(f"  โ ะขะพัะฝะพััั ะผะพะดะตะปะธ ั ะพัะพะฑัะฐะฝะฝัะผะธ ะฟัะธะทะฝะฐะบะฐะผะธ: {score:.4f}")
    
    return score, top_features.index
```

ะฟะพะดะฑะพั ะณะธะฟะตัะฟะฐัะฐะผะตััะพะฒ (ะฟะพะดะฑะพั ะณะธะฟะตัะฟะฐัะฐะผะตััะพะฒ ั ะธัะฟะพะปัะทะพะฒะฐะฝะธะตะผ ะบัะพัั-ะฒะฐะปะธะดะฐัะธะธ)

```python
def hyperparameter_tuning_cv(X, y, problem_type='classification', dataset_name="ะะฐะฝะฝัะต"):
    """ะะพะดะฑะพั ะณะธะฟะตัะฟะฐัะฐะผะตััะพะฒ ั ะธัะฟะพะปัะทะพะฒะฐะฝะธะตะผ ะบัะพัั-ะฒะฐะปะธะดะฐัะธะธ"""
    print(f"\n๐ฏ ะะะะะะ ะะะะะะะะะะะะขะะะ ะก ะะะะกะก-ะะะะะะะฆะะะ ะะะฏ {dataset_name}")
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, 
        stratify=y if problem_type == 'classification' else None
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    if problem_type == 'classification':
        model = KNeighborsClassifier()
        scoring = 'accuracy'
        param_grid = {
            'n_neighbors': range(1, 31),
            'weights': ['uniform', 'distance'],
            'metric': ['euclidean', 'manhattan', 'minkowski'],
            'p': [1, 2] 
        }
    else:
        model = KNeighborsRegressor()
        scoring = 'r2'
        param_grid = {
            'n_neighbors': range(1, 31),
            'weights': ['uniform', 'distance'],
            'metric': ['euclidean', 'manhattan', 'minkowski'],
            'p': [1, 2]
        }
    
    if problem_type == 'classification':
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    else:
        cv = KFold(n_splits=5, shuffle=True, random_state=42)
    
    print("๐ ะะฐะฟััะบ GridSearchCV...")
    grid_search = GridSearchCV(
        model, param_grid, cv=cv, scoring=scoring, 
        n_jobs=-1, verbose=0, return_train_score=True
    )
    
    grid_search.fit(X_train_scaled, y_train)
    
    best_params = grid_search.best_params_
    best_score = grid_search.best_score_
    best_model = grid_search.best_estimator_
    
    test_score = best_model.score(X_test_scaled, y_test)
    
    print(f"  โ ะัััะธะต ะฟะฐัะฐะผะตััั: {best_params}")
    print(f"  โ ะัััะฐั ะพัะตะฝะบะฐ ะฝะฐ ะบัะพัั-ะฒะฐะปะธะดะฐัะธะธ: {best_score:.4f}")
    print(f"  โ ะัะตะฝะบะฐ ะฝะฐ ัะตััะพะฒะพะน ะฒัะฑะพัะบะต: {test_score:.4f}")
    
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    neighbors_results = {}
    
    for n in param_grid['n_neighbors']:
        mask = grid_search.cv_results_['param_n_neighbors'] == n
        if mask.any():
            neighbors_results[n] = {
                'mean_train': grid_search.cv_results_['mean_train_score'][mask].mean(),
                'mean_test': grid_search.cv_results_['mean_test_score'][mask].mean(),
                'std_test': grid_search.cv_results_['std_test_score'][mask].mean()
            }
    
    neighbors = list(neighbors_results.keys())
    train_scores = [neighbors_results[n]['mean_train'] for n in neighbors]
    test_scores = [neighbors_results[n]['mean_test'] for n in neighbors]
    test_stds = [neighbors_results[n]['std_test'] for n in neighbors]
    
    plt.plot(neighbors, train_scores, 'o-', label='Train Score', linewidth=2)
    plt.plot(neighbors, test_scores, 's-', label='CV Test Score', linewidth=2)
    plt.fill_between(neighbors, 
                    np.array(test_scores) - np.array(test_stds),
                    np.array(test_scores) + np.array(test_stds),
                    alpha=0.2)
    plt.axvline(x=best_params['n_neighbors'], color='red', linestyle='--', 
                label=f'ะัััะตะต k: {best_params["n_neighbors"]}')
    plt.xlabel('ะะพะปะธัะตััะฒะพ ัะพัะตะดะตะน (k)')
    plt.ylabel('Score')
    plt.title('ะะปะธัะฝะธะต n_neighbors ะฝะฐ ัะพัะฝะพััั')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 2)
    metric_results = {}
    
    for metric in param_grid['metric']:
        mask = grid_search.cv_results_['param_metric'] == metric
        if mask.any():
            metric_results[metric] = grid_search.cv_results_['mean_test_score'][mask].max()
    
    metrics = list(metric_results.keys())
    metric_scores = list(metric_results.values())
    
    bars = plt.bar(metrics, metric_scores, alpha=0.7, color=['skyblue', 'lightgreen', 'salmon'])
    plt.ylabel('Score')
    plt.title('ะกัะฐะฒะฝะตะฝะธะต ะผะตััะธะบ ัะฐัััะพัะฝะธั')
    
    for bar, score in zip(bars, metric_scores):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
                f'{score:.4f}', ha='center', va='bottom')
    
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 3)
    weight_results = {}
    
    for weight in param_grid['weights']:
        mask = grid_search.cv_results_['param_weights'] == weight
        if mask.any():
            weight_results[weight] = grid_search.cv_results_['mean_test_score'][mask].max()
    
    weights = list(weight_results.keys())
    weight_scores = list(weight_results.values())
    
    bars = plt.bar(weights, weight_scores, alpha=0.7, color=['gold', 'lightcoral'])
    plt.ylabel('Score')
    plt.title('ะกัะฐะฒะฝะตะฝะธะต ะฒะตัะพะฒ ัะพัะตะดะตะน')
    
    for bar, score in zip(bars, weight_scores):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
                f'{score:.4f}', ha='center', va='bottom')
    
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    train_test_gap = neighbors_results[best_params['n_neighbors']]['mean_train'] - best_score
    print(f"\n๐ ะะะะะะ ะะะะะะะฃะงะะะะฏ:")
    print(f"  ะะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ CV score: {train_test_gap:.4f}")
    
    if train_test_gap > 0.1:
        print("  โ๏ธ ะะพะทะผะพะถะฝะพะต ะฟะตัะตะพะฑััะตะฝะธะต (ะฑะพะปััะฐั ัะฐะทะฝะธัะฐ)")
    else:
        print("  โ ะฅะพัะพัะตะต ะพะฑะพะฑัะตะฝะธะต (ะผะฐะปะฐั ัะฐะทะฝะธัะฐ)")
    
    return best_model, best_params, test_score
```

ะพะฟัะธะผะธะทะฐัะธั ะฟะพัะพะณะฐ ะบะปะฐััะธัะธะบะฐัะธะธ (ะฝะฐัััะพะนะบะฐ ะฟะพัะพะณะฐ ะบะปะฐััะธัะธะบะฐัะธะธ ัะปัััะธั precision/recall)

```python
def optimize_threshold(model, X_test, y_test, model_name="ะะพะดะตะปั"):
    """ะะฟัะธะผะธะทะฐัะธั ะฟะพัะพะณะฐ ะบะปะฐััะธัะธะบะฐัะธะธ ะดะปั ัะปัััะตะฝะธั ะผะตััะธะบ"""
    from sklearn.metrics import precision_recall_curve, classification_report, accuracy_score, f1_score
    
    print(f"\n๐ฏ ะะะขะะะะะะฆะะฏ ะะะะะะ ะะะะกะกะะคะะะะฆะะ ะะะฏ {model_name}")
    
    y_proba = model.predict_proba(X_test)[:, 1]
    precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
    
    f1_scores = 2 * (precision[:-1] * recall[:-1]) / (precision[:-1] + recall[:-1] + 1e-8)
    
    best_f1_index = np.argmax(f1_scores)
    best_threshold = thresholds[best_f1_index]
    best_f1 = f1_scores[best_f1_index]
    
    balance_index = np.argmin(np.abs(precision[:-1] - recall[:-1]))
    balance_threshold = thresholds[balance_index]
    balance_f1 = f1_scores[balance_index]
    
    print(f"  โ ะะฟัะธะผะฐะปัะฝัะน ะฟะพัะพะณ ะฟะพ F1-score: {best_threshold:.4f} (F1 = {best_f1:.4f})")
    print(f"  โ ะกะฑะฐะปะฐะฝัะธัะพะฒะฐะฝะฝัะน ะฟะพัะพะณ: {balance_threshold:.4f} (F1 = {balance_f1:.4f})")
    
    y_pred_default = (y_proba >= 0.5).astype(int)
    default_accuracy = accuracy_score(y_test, y_pred_default)
    default_f1 = f1_score(y_test, y_pred_default)
    
    y_pred_optimized = (y_proba >= best_threshold).astype(int)
    optimized_accuracy = accuracy_score(y_test, y_pred_optimized)
    optimized_f1 = f1_score(y_test, y_pred_optimized)
    
    print(f"\n๐ ะกะะะะะะะะ ะก ะะะะะะะ ะะ ะฃะะะะงะะะะฎ (0.5):")
    print(f"  ะะพ ัะผะพะปัะฐะฝะธั (0.5):    Accuracy = {default_accuracy:.4f}, F1 = {default_f1:.4f}")
    print(f"  ะะฟัะธะผะธะทะธัะพะฒะฐะฝะฝัะน:      Accuracy = {optimized_accuracy:.4f}, F1 = {optimized_f1:.4f}")
    
    improvement_f1 = optimized_f1 - default_f1
    if improvement_f1 > 0:
        print(f"  ๐ ะฃะะฃะงะจะะะะ F1: +{improvement_f1:.4f}")
    else:
        print(f"  ๐ ะฃะฅะฃะะจะะะะ F1: {improvement_f1:.4f}")
    
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    plt.plot(thresholds, precision[:-1], 'b-', label='Precision', linewidth=2)
    plt.plot(thresholds, recall[:-1], 'g-', label='Recall', linewidth=2)
    plt.plot(thresholds, f1_scores, 'r-', label='F1-Score', linewidth=2)
    plt.axvline(x=best_threshold, color='red', linestyle='--', 
                label=f'ะัััะธะน ะฟะพัะพะณ: {best_threshold:.3f}', alpha=0.7)
    plt.axvline(x=0.5, color='black', linestyle='--', 
                label='ะะพัะพะณ ะฟะพ ัะผะพะปัะฐะฝะธั: 0.5', alpha=0.7)
    plt.axvline(x=balance_threshold, color='orange', linestyle='--', 
                label=f'ะกะฑะฐะปะฐะฝัะธัะพะฒะฐะฝะฝัะน: {balance_threshold:.3f}', alpha=0.7)
    plt.xlabel('ะะพัะพะณ ะบะปะฐััะธัะธะบะฐัะธะธ')
    plt.ylabel('Score')
    plt.title('Precision, Recall ะธ F1 vs ะะพัะพะณ')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 2)
    plt.plot(recall, precision, 'b-', linewidth=2)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('ะัะธะฒะฐั Precision-Recall')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 3)
    metrics = ['Accuracy', 'F1-Score']
    default_scores = [default_accuracy, default_f1]
    optimized_scores = [optimized_accuracy, optimized_f1]
    
    x = np.arange(len(metrics))
    width = 0.35
    
    plt.bar(x - width/2, default_scores, width, label='ะะพัะพะณ 0.5', alpha=0.7)
    plt.bar(x + width/2, optimized_scores, width, label='ะะฟัะธะผะธะทะธัะพะฒะฐะฝะฝัะน', alpha=0.7)
    
    plt.xlabel('ะะตััะธะบะธ')
    plt.ylabel('Score')
    plt.title('ะกัะฐะฒะฝะตะฝะธะต ะผะตััะธะบ\nะดะพ ะธ ะฟะพัะปะต ะพะฟัะธะผะธะทะฐัะธะธ ะฟะพัะพะณะฐ')
    plt.xticks(x, metrics)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    for i, (default, optimized) in enumerate(zip(default_scores, optimized_scores)):
        plt.text(i - width/2, default + 0.01, f'{default:.3f}', ha='center', va='bottom')
        plt.text(i + width/2, optimized + 0.01, f'{optimized:.3f}', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n๐ ะะะขะะะฌะะซะ ะะขะงะะข ะะะะกะกะะคะะะะฆะะ:")
    print("  ะก ะฟะพัะพะณะพะผ ะฟะพ ัะผะพะปัะฐะฝะธั (0.5):")
    print(classification_report(y_test, y_pred_default, target_names=['ะะตััะฟะตั', 'ะฃัะฟะตั']))
    
    print("  ะก ะพะฟัะธะผะธะทะธัะพะฒะฐะฝะฝัะผ ะฟะพัะพะณะพะผ:")
    print(classification_report(y_test, y_pred_optimized, target_names=['ะะตััะฟะตั', 'ะฃัะฟะตั']))
    
    return best_threshold, optimized_f1, optimized_accuracy
```

```python
from sklearn.metrics import classification_report, confusion_matrix, mean_squared_error, r2_score, accuracy_score, mean_absolute_error, precision_recall_curve, f1_score
from sklearn.model_selection import GridSearchCV, StratifiedKFold, KFold, train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def final_classification_model(df_student):
    """ะคะธะฝะฐะปัะฝะฐั ะผะพะดะตะปั ะบะปะฐััะธัะธะบะฐัะธะธ ั ัะปัััะตะฝะฝัะผ ะฑะตะนะทะปะฐะนะฝะพะผ"""
    print("\n" + "="*80)
    print("๐ฏ ะคะะะะะฌะะะฏ ะะะะะะฌ ะะะะกะกะะคะะะะฆะะ")
    print("ะะธะฟะพัะตะทั: ะัะตะฟัะพัะตััะธะฝะณ ะดะฐะฝะฝัั + ะะฟัะธะผะธะทะฐัะธั ะฟะพัะพะณะฐ")
    print("="*80)
    
    print("๐ง 1. ะะะะะะะะะะ ะฃะะฃะงะจะะะะะะ ะะะะะะะฆะะกะกะะะะ...")
    df_processed = advanced_preprocessing(df_student)
    X, y, features = preprocess_student_classification(df_processed)
    
    if X is None:
        print("โ ะัะธะฑะบะฐ ะฟัะธ ะฟะพะดะณะพัะพะฒะบะต ะดะฐะฝะฝัั")
        return None, None, None
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    if 'best_params_student' in locals():
        knn_model = KNeighborsClassifier(**best_params_student)
        print(f"๐ง ะัะฟะพะปัะทััััั ะปัััะธะต ะฟะฐัะฐะผะตััั: {best_params_student}")
    else:
        knn_model = KNeighborsClassifier(n_neighbors=5)
        print("๐ง ะัะฟะพะปัะทััััั ะฟะฐัะฐะผะตััั ะฟะพ ัะผะพะปัะฐะฝะธั")
    
    knn_model.fit(X_train_scaled, y_train)
    
    y_pred_default = knn_model.predict(X_test_scaled)
    default_accuracy = accuracy_score(y_test, y_pred_default)
    
    print(f"\n๐ ะะะะฃะะฌะขะะขะซ ะก ะะะะะะะ ะะ ะฃะะะะงะะะะฎ (0.5):")
    print(f"  ะขะพัะฝะพััั: {default_accuracy:.4f}")
    print(classification_report(y_test, y_pred_default, target_names=['ะะตััะฟะตั', 'ะฃัะฟะตั']))
    
    if hasattr(knn_model, 'predict_proba'):
        print("\n๐ฏ 2. ะะะขะะะะะะฆะะฏ ะะะะะะ ะะะะกะกะะคะะะะฆะะ...")
        
        y_proba = knn_model.predict_proba(X_test_scaled)[:, 1]
        precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
        
        f1_scores = 2 * (precision[:-1] * recall[:-1]) / (precision[:-1] + recall[:-1] + 1e-8)
        
        best_f1_index = np.argmax(f1_scores)
        best_threshold = thresholds[best_f1_index]
        best_f1 = f1_scores[best_f1_index]
        
        print(f"  โ ะะฟัะธะผะฐะปัะฝัะน ะฟะพัะพะณ ะฟะพ F1-score: {best_threshold:.4f} (F1 = {best_f1:.4f})")
        
        y_pred_optimized = (y_proba >= best_threshold).astype(int)
        optimized_accuracy = accuracy_score(y_test, y_pred_optimized)
        optimized_f1 = f1_score(y_test, y_pred_optimized)
        
        print(f"\n๐ ะคะะะะะฌะะซะ ะะะะฃะะฌะขะะขะซ ะก ะะะขะะะะะะะะะะะะซะ ะะะะะะะ:")
        print(f"  ะขะพัะฝะพััั: {optimized_accuracy:.4f}")
        print(f"  F1-Score: {optimized_f1:.4f}")
        
        improvement = optimized_accuracy - default_accuracy
        if improvement > 0:
            print(f"  ๐ ะฃะะฃะงะจะะะะ ะขะะงะะะกะขะ: +{improvement:.4f} (+{improvement/default_accuracy*100:.2f}%)")
        else:
            print(f"  ๐ ะฃะฅะฃะะจะะะะ ะขะะงะะะกะขะ: {improvement:.4f}")
        
        plt.figure(figsize=(15, 5))
        
        plt.subplot(1, 3, 1)
        plt.plot(thresholds, precision[:-1], 'b-', label='Precision', linewidth=2)
        plt.plot(thresholds, recall[:-1], 'g-', label='Recall', linewidth=2)
        plt.plot(thresholds, f1_scores, 'r-', label='F1-Score', linewidth=2)
        plt.axvline(x=best_threshold, color='red', linestyle='--', 
                    label=f'ะัััะธะน ะฟะพัะพะณ: {best_threshold:.3f}', alpha=0.7)
        plt.axvline(x=0.5, color='black', linestyle='--', 
                    label='ะะพัะพะณ ะฟะพ ัะผะพะปัะฐะฝะธั: 0.5', alpha=0.7)
        plt.xlabel('ะะพัะพะณ ะบะปะฐััะธัะธะบะฐัะธะธ')
        plt.ylabel('Score')
        plt.title('Precision, Recall ะธ F1 vs ะะพัะพะณ')
        plt.legend()
        plt.grid(True, alpha=0.3)
        
        plt.subplot(1, 3, 2)
        methods = ['ะะพัะพะณ 0.5', 'ะะฟัะธะผะธะทะธัะพะฒะฐะฝะฝัะน']
        scores = [default_accuracy, optimized_accuracy]
        colors = ['lightblue', 'lightgreen']
        
        bars = plt.bar(methods, scores, color=colors, alpha=0.7)
        plt.ylabel('ะขะพัะฝะพััั (Accuracy)')
        plt.title('ะกัะฐะฒะฝะตะฝะธะต ัะพัะฝะพััะธ')
        plt.ylim(0, 1)
        
        for bar, score in zip(bars, scores):
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
                    f'{score:.4f}', ha='center', va='bottom')
        
        plt.subplot(1, 3, 3)
        cm_default = confusion_matrix(y_test, y_pred_default)
        cm_optimized = confusion_matrix(y_test, y_pred_optimized)
        
        plt.text(0.1, 0.7, f'ะะพัะพะณ 0.5:\nะขะพัะฝะพััั: {default_accuracy:.4f}\n\n{cm_default}', 
                transform=plt.gca().transAxes, fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.7))
        plt.text(0.5, 0.7, f'ะะพัะพะณ {best_threshold:.3f}:\nะขะพัะฝะพััั: {optimized_accuracy:.4f}\n\n{cm_optimized}', 
                transform=plt.gca().transAxes, fontsize=10, bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.7))
        
        plt.axis('off')
        plt.title('ะะฐััะธัั ะพัะธะฑะพะบ')
        
        plt.tight_layout()
        plt.show()
        
        print("\n๐ ะคะะะะะฌะะซะ ะะขะงะะข ะะะะกะกะะคะะะะฆะะ:")
        print(classification_report(y_test, y_pred_optimized, target_names=['ะะตััะฟะตั', 'ะฃัะฟะตั']))
        
        return knn_model, best_threshold, optimized_accuracy
    
    else:
        print("โ ะะพะดะตะปั ะฝะต ะฟะพะดะดะตัะถะธะฒะฐะตั ะพะฟัะธะผะธะทะฐัะธั ะฟะพัะพะณะฐ")
        return knn_model, 0.5, default_accuracy
    
def final_regression_model(df_house):
    """ะคะธะฝะฐะปัะฝะฐั ะผะพะดะตะปั ัะตะณัะตััะธะธ ั ัะปัััะตะฝะฝัะผ ะฑะตะนะทะปะฐะนะฝะพะผ"""
    print("\n" + "="*80)
    print("๐ฏ ะคะะะะะฌะะะฏ ะะะะะะฌ ะะะะะะกะกะะ")
    print("ะะธะฟะพัะตะทั: ะัะตะฟัะพัะตััะธะฝะณ ะดะฐะฝะฝัั + ะะพะดะฑะพั ะณะธะฟะตัะฟะฐัะฐะผะตััะพะฒ")
    print("="*80)
    
    print("๐ง 1. ะะะะะะะะะะ ะฃะะฃะงะจะะะะะะ ะะะะะะะฆะะกะกะะะะ...")
    df_processed = advanced_preprocessing_house(df_house)
    X, y, features = preprocess_house_regression(df_processed)
    
    if X is None:
        print("โ ะัะธะฑะบะฐ ะฟัะธ ะฟะพะดะณะพัะพะฒะบะต ะดะฐะฝะฝัั")
        return None, None, None
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    print("\n๐ฏ 2. ะะะะะะ ะะะะะะะะะะะะขะะะ ะะ ะะะะกะก-ะะะะะะะฆะะ...")

    model = KNeighborsRegressor()
    param_grid = {
        'n_neighbors': range(3, 20, 2),
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    }
    
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    
    grid_search = GridSearchCV(
        model, param_grid, cv=cv, scoring='r2', 
        n_jobs=-1, verbose=0
    )
    
    grid_search.fit(X_train_scaled, y_train)
    
    best_model = grid_search.best_estimator_
    best_params = grid_search.best_params_
    best_score = grid_search.best_score_
    
    print(f"  โ ะัััะธะต ะฟะฐัะฐะผะตััั: {best_params}")
    print(f"  โ ะัััะธะน Rยฒ ะฝะฐ ะบัะพัั-ะฒะฐะปะธะดะฐัะธะธ: {best_score:.4f}")
    
    y_pred = best_model.predict(X_test_scaled)
    test_r2 = r2_score(y_test, y_pred)
    test_mse = mean_squared_error(y_test, y_pred)
    test_rmse = np.sqrt(test_mse)
    test_mae = mean_absolute_error(y_test, y_pred)
    
    print(f"\n๐ ะคะะะะะฌะะซะ ะะะะฃะะฌะขะะขะซ ะะ ะขะะกะขะะะะ ะะซะะะะะ:")
    print(f"  Rยฒ Score: {test_r2:.4f}")
    print(f"  MSE: {test_mse:.2f}")
    print(f"  RMSE: {test_rmse:.2f}")
    print(f"  MAE: {test_mae:.2f}")
    
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    plt.scatter(y_test, y_pred, alpha=0.6, s=50)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
    plt.xlabel('ะคะฐะบัะธัะตัะบะธะต ะทะฝะฐัะตะฝะธั')
    plt.ylabel('ะัะตะดัะบะฐะทะฐะฝะฝัะต ะทะฝะฐัะตะฝะธั')
    plt.title(f'ะัะตะดัะบะฐะทะฐะฝะธั vs ะคะฐะบัะธัะตัะบะธะต ะทะฝะฐัะตะฝะธั\nRยฒ = {test_r2:.4f}')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 2)
    residuals = y_test - y_pred
    plt.scatter(y_pred, residuals, alpha=0.6, s=50)
    plt.axhline(y=0, color='red', linestyle='--')
    plt.xlabel('ะัะตะดัะบะฐะทะฐะฝะฝัะต ะทะฝะฐัะตะฝะธั')
    plt.ylabel('ะััะฐัะบะธ')
    plt.title('ะะฝะฐะปะธะท ะพััะฐัะบะพะฒ')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 3)
    plt.hist(residuals, bins=30, alpha=0.7, edgecolor='black')
    plt.xlabel('ะัะธะฑะบะฐ')
    plt.ylabel('ะงะฐััะพัะฐ')
    plt.title('ะะฐัะฟัะตะดะตะปะตะฝะธะต ะพัะธะฑะพะบ ะฟัะตะดัะบะฐะทะฐะฝะธั')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print("\n๐ ะะะะะะ ะะะงะะกะขะะ ะะะะะะ:")
    if test_r2 >= 0.7:
        print("  ๐ ะะขะะะงะะะ ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (Rยฒ โฅ 0.7)")
    elif test_r2 >= 0.5:
        print("  ๐ ะฅะะะะจะะ ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.5 โค Rยฒ < 0.7)")
    elif test_r2 >= 0.3:
        print("  ๐ ะฃะะะะะะขะะะะะขะะะฌะะะ ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.3 โค Rยฒ < 0.5)")
    else:
        print("  ๐ก ะะะะะะ ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (Rยฒ < 0.3)")
    
    return best_model, best_params, test_r2
```
