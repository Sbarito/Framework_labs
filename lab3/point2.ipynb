```python
def preprocess_student_classification(df):
    """ะัะตะดะพะฑัะฐะฑะพัะบะฐ ัััะดะตะฝัะตัะบะธั ะดะฐะฝะฝัั ะดะปั ะบะปะฐััะธัะธะบะฐัะธะธ"""
    print("\n" + "="*50)
    print("ะะะะะะะะะะะขะะ ะกะขะฃะะะะงะะกะะะฅ ะะะะะซะฅ (ะะะะกะกะะคะะะะฆะะฏ)")
    print("="*50)
    
    if df is None:
        print("โ DataFrame ะฟัััะพะน")
        return None, None, None
    
    df_clean = df.copy()
    
    print("ะะพัััะฟะฝัะต ะบะพะปะพะฝะบะธ:")
    for i, col in enumerate(df_clean.columns):
        print(f"  {i+1:2d}. '{col}' (ัะธะฟ: {df_clean[col].dtype})")
    
    if 'G3' not in df_clean.columns:
        print("โ ะะพะปะพะฝะบะฐ 'G3' ะฝะต ะฝะฐะนะดะตะฝะฐ!")
        print("   ะะพัััะฟะฝัะต ะบะพะปะพะฝะบะธ:", list(df_clean.columns))
        return None, None, None
    
    print(f"โ ะะฐะนะดะตะฝะฐ ัะตะปะตะฒะฐั ะฟะตัะตะผะตะฝะฝะฐั: 'G3'")
    print(f"  ะะธะฐะฟะฐะทะพะฝ ะทะฝะฐัะตะฝะธะน: {df_clean['G3'].min()} - {df_clean['G3'].max()}")
    
    df_clean['academic_success'] = (df_clean['G3'] >= 10).astype(int)
    success_rate = df_clean['academic_success'].mean() * 100
    print(f"โ ะกะพะทะดะฐะฝะฐ ัะตะปะตะฒะฐั ะฟะตัะตะผะตะฝะฝะฐั 'academic_success'")
    print(f"  ะัะพัะตะฝั ััะฟะตะฒะฐััะธั ัััะดะตะฝัะพะฒ: {success_rate:.1f}%")
    
    numeric_features = []
    possible_numeric = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 
                       'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']
    
    print("\nะะพะธัะบ ัะธัะปะพะฒัั ะฟัะธะทะฝะฐะบะพะฒ:")
    for feature in possible_numeric:
        if feature in df_clean.columns:
            if df_clean[feature].dtype == 'object':
                df_clean[feature] = pd.to_numeric(df_clean[feature], errors='coerce')
            numeric_features.append(feature)
            print(f"  โ '{feature}' ะฝะฐะนะดะตะฝ")
        else:
            print(f"  โ '{feature}' ะฝะต ะฝะฐะนะดะตะฝ")
    
    categorical_features = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 
                           'Fjob', 'reason', 'guardian']
    
    features = numeric_features.copy()
    le = LabelEncoder()
    
    print("\nะะพะธัะบ ะธ ะบะพะดะธัะพะฒะฐะฝะธะต ะบะฐัะตะณะพัะธะฐะปัะฝัั ะฟัะธะทะฝะฐะบะพะฒ:")
    for feature in categorical_features:
        if feature in df_clean.columns:
            try:
                df_clean[f'{feature}_encoded'] = le.fit_transform(df_clean[feature].astype(str))
                features.append(f'{feature}_encoded')
                print(f"  โ '{feature}' ะทะฐะบะพะดะธัะพะฒะฐะฝ")
            except Exception as e:
                print(f"  โ '{feature}' ะพัะธะฑะบะฐ ะบะพะดะธัะพะฒะฐะฝะธั: {e}")
        else:
            print(f"  โ '{feature}' ะฝะต ะฝะฐะนะดะตะฝ")
    
    initial_count = len(df_clean)
    df_clean = df_clean.dropna(subset=features + ['academic_success'])
    final_count = len(df_clean)
    
    print(f"\nโ ะัะธััะบะฐ ะดะฐะฝะฝัั: {initial_count} โ {final_count} ัััะพะบ")
    print(f"โ ะัะฟะพะปัะทัะตััั {len(features)} ะฟัะธะทะฝะฐะบะพะฒ")
    
    if not features or df_clean.empty:
        print("โ ะะตะดะพััะฐัะพัะฝะพ ะดะฐะฝะฝัั ะดะปั ะฐะฝะฐะปะธะทะฐ")
        return None, None, None
    
    X = df_clean[features]
    y = df_clean['academic_success']
    
    print(f"โ ะัะพะณะพะฒัะน ะฝะฐะฑะพั ะดะฐะฝะฝัั: {X.shape[0]} samples, {X.shape[1]} features")
    print(f"  ะะฐะปะฐะฝั ะบะปะฐััะพะฒ: {np.bincount(y)}")
    
    return X, y, features
```

ะคัะฝะบัะธะธ ะฟัะตะดะพะฑัะฐะฑะพัะบะธ ะดะฐะฝะฝัั ะดะปั ัะตะณัะตััะธะธ (ะดะฐะฝะฝัะต ะพ ะฝะตะดะฒะธะถะธะผะพััะธ)

```python
def preprocess_house_regression(df):
    """ะัะตะดะพะฑัะฐะฑะพัะบะฐ ะดะฐะฝะฝัั ะพ ะฝะตะดะฒะธะถะธะผะพััะธ ะดะปั ัะตะณัะตััะธะธ"""
    print("\n" + "="*50)
    print("ะะะะะะะะะะะขะะ ะะะะะซะฅ ะ ะะะะะะะะะะกะขะ (ะะะะะะกะกะะฏ)")
    print("="*50)
    
    if df is None:
        return None, None, None
    
    df_clean = df.copy()
    
    print(f"โ ะฆะตะปะตะฒะฐั ะฟะตัะตะผะตะฝะฝะฐั: Rent (ะฐัะตะฝะดะฝะฐั ะฟะปะฐัะฐ)")
    print(f"  ะะธะฐะฟะฐะทะพะฝ ัะตะฝ: {df_clean['Rent'].min()} - {df_clean['Rent'].max()}")
    
    if 'Floor' in df_clean.columns:
        def parse_floor(floor_str):
            if pd.isna(floor_str):
                return 0
            floor_str = str(floor_str)
            if 'Ground' in floor_str:
                return 0
            if 'Basement' in floor_str:
                return -1
            try:
                parts = floor_str.split()
                for part in parts:
                    if part.isdigit():
                        return int(part)
                return 0
            except:
                return 0
        
        df_clean['floor_number'] = df_clean['Floor'].apply(parse_floor)
        print("โ ะะฑัะฐะฑะพัะฐะฝะฐ ะบะพะปะพะฝะบะฐ Floor")
    
    if 'Size' in df_clean.columns:
        try:
            df_clean['Size'] = df_clean['Size'].astype(str).str.replace(' sqft', '').astype(float)
            print("โ ะะฑัะฐะฑะพัะฐะฝะฐ ะบะพะปะพะฝะบะฐ Size")
        except Exception as e:
            print(f"โ ะัะธะฑะบะฐ ะพะฑัะฐะฑะพัะบะธ Size: {e}")
    
    features = ['BHK', 'Bathroom']
    if 'Size' in df_clean.columns:
        features.append('Size')
    if 'floor_number' in df_clean.columns:
        features.append('floor_number')
    

    categorical_cols = ['Area Type', 'City', 'Furnishing Status', 'Tenant Preferred', 'Point of Contact']
    le = LabelEncoder()
    
    for col in categorical_cols:
        if col in df_clean.columns:
            try:
                df_clean[col + '_encoded'] = le.fit_transform(df_clean[col].astype(str))
                features.append(col + '_encoded')
                print(f"โ ะะฐะบะพะดะธัะพะฒะฐะฝะฐ ะบะพะปะพะฝะบะฐ: {col}")
            except Exception as e:
                print(f"โ ะัะธะฑะบะฐ ะบะพะดะธัะพะฒะฐะฝะธั {col}: {e}")
    
    X = df_clean[features]
    y = df_clean['Rent']
    
    print(f"โ ะัะพะณะพะฒัะน ะฝะฐะฑะพั ะดะฐะฝะฝัั: {X.shape[0]} samples, {X.shape[1]} features")
    print(f"  ะกัะฐัะธััะธะบะฐ ะฐัะตะฝะดะฝะพะน ะฟะปะฐัั: mean={y.mean():.2f}, std={y.std():.2f}")
    
    return X, y, features
```

ะคัะฝะบัะธั ะธััะปะตะดะพะฒะฐะฝะธั c ัะตัะฐััะธะผ ะดะตัะฒะพะผ ะดะปั ะบะปะฐััะธัะธะบะฐัะธะธ

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, classification_report

def decision_tree_classification_study(X, y, dataset_name="ะะฐะฝะฝัะต"):
    print(f"\n" + "="*70)
    print(f"ะะะจะะฎะฉะะ ะะะะะะ ะะะะกะกะะคะะะะฆะะฏ - {dataset_name}")
    print("="*70)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )
    
    print("๐ ะััะปะตะดะพะฒะฐะฝะธะต ะฒะปะธัะฝะธั ะณะปัะฑะธะฝั ะดะตัะตะฒะฐ ะฝะฐ ัะพัะฝะพััั...")
    depth_range = range(1, 21)
    train_scores = []
    test_scores = []
    
    for depth in depth_range:
        tree = DecisionTreeClassifier(max_depth=depth, random_state=42)
        tree.fit(X_train, y_train)
        train_scores.append(tree.score(X_train, y_train))
        test_scores.append(tree.score(X_test, y_test))
    
    best_depth_index = np.argmax(test_scores)
    best_depth = depth_range[best_depth_index]
    best_score = test_scores[best_depth_index]
    best_train_score = train_scores[best_depth_index]
    
    print(f"๐ฏ ะะตะทัะปััะฐัั:")
    print(f"  ะะฟัะธะผะฐะปัะฝะฐั ะณะปัะฑะธะฝะฐ: {best_depth}")
    print(f"  ะัััะฐั ัะพัะฝะพััั: {best_score:.4f}")
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(depth_range, train_scores, label='ะะฑััะฐััะฐั ะฒัะฑะพัะบะฐ', marker='o', linewidth=2)
    plt.plot(depth_range, test_scores, label='ะขะตััะพะฒะฐั ะฒัะฑะพัะบะฐ', marker='s', linewidth=2)
    plt.axvline(x=best_depth, color='red', linestyle='--', alpha=0.7, label=f'ะัััะฐั ะณะปัะฑะธะฝะฐ={best_depth}')
    plt.xlabel('ะะฐะบัะธะผะฐะปัะฝะฐั ะณะปัะฑะธะฝะฐ ะดะตัะตะฒะฐ')
    plt.ylabel('ะขะพัะฝะพััั')
    plt.title(f'ะะตัะฐััะตะต ะดะตัะตะฒะพ - ะะปะธัะฝะธะต ะณะปัะฑะธะฝั\n{dataset_name}', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    best_tree = DecisionTreeClassifier(max_depth=best_depth, random_state=42)
    best_tree.fit(X_train, y_train)
    y_pred = best_tree.predict(X_test)
    
    plt.subplot(1, 2, 2)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['ะะตััะฟะตั', 'ะฃัะฟะตั'], 
                yticklabels=['ะะตััะฟะตั', 'ะฃัะฟะตั'])
    plt.title(f'ะะฐััะธัะฐ ะพัะธะฑะพะบ (ะณะปัะฑะธะฝะฐ={best_depth})\nะขะพัะฝะพััั: {best_score:.4f}', fontsize=12)
    plt.ylabel('ะคะฐะบัะธัะตัะบะธะต ะทะฝะฐัะตะฝะธั')
    plt.xlabel('ะัะตะดัะบะฐะทะฐะฝะฝัะต ะทะฝะฐัะตะฝะธั')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n๐ ะะะะะะ ะะะะะะ ะะะจะะฎะฉะะะ ะะะะะะ:")
    print(f"  ะะฟัะธะผะฐะปัะฝะฐั ะณะปัะฑะธะฝะฐ ะดะตัะตะฒะฐ: {best_depth}")
    print(f"  ะขะพัะฝะพััั ะฝะฐ ัะตััะพะฒะพะน ะฒัะฑะพัะบะต: {best_score:.4f}")
    print(f"  ะขะพัะฝะพััั ะฝะฐ ะพะฑััะฐััะตะน ะฒัะฑะพัะบะต: {best_train_score:.4f}")
    
    overfitting_gap = best_train_score - best_score
    print(f"  ะะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test accuracy: {overfitting_gap:.4f}")
    
    if overfitting_gap > 0.1:
        print("  โ๏ธ ะะพะทะผะพะถะฝะพะต ะฟะตัะตะพะฑััะตะฝะธะต (ะฑะพะปััะฐั ัะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test)")
    else:
        print("  โ ะะพะดะตะปั ัะพัะพัะพ ะพะฑะพะฑัะฐะตั (ะผะฐะปะฐั ัะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test)")
    
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    print(f"\n๐ ะะะะะะะะขะะะฌะะซะ ะะะขะะะะ ะะะะกะกะะคะะะะฆะะ:")
    print(f"  Precision (ะขะพัะฝะพััั): {precision:.4f}")
    print(f"  Recall (ะะพะปะฝะพัะฐ): {recall:.4f}")
    print(f"  F1-Score: {f1:.4f}")
    
    if best_score >= 0.9:
        print("  ๐ ะัะปะธัะฝะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (ะขะพัะฝะพััั โฅ 0.9)")
    elif best_score >= 0.8:
        print("  ๐ ะฅะพัะพัะตะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.8 โค ะขะพัะฝะพััั < 0.9)")
    elif best_score >= 0.7:
        print("  ๐ ะฃะดะพะฒะปะตัะฒะพัะธัะตะปัะฝะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.7 โค ะขะพัะฝะพััั < 0.8)")
    else:
        print("  ๐ก ะะธะทะบะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (ะขะพัะฝะพััั < 0.7)")
    
    print(f"\n๐ ะะะขะะะฌะะซะ ะะขะงะะข ะะ ะะะะกะกะะ:")
    print(classification_report(y_test, y_pred, target_names=['ะะตััะฟะตั', 'ะฃัะฟะตั']))
    
    print(f"\n๐ณ ะะะะะะกะขะฌ ะะะะะะะะะ:")
    feature_importances = best_tree.feature_importances_
    feature_names = X.columns if hasattr(X, 'columns') else [f'ะัะธะทะฝะฐะบ {i}' for i in range(X.shape[1])]
    
    importance_df = pd.DataFrame({
        'ะัะธะทะฝะฐะบ': feature_names,
        'ะะฐะถะฝะพััั': feature_importances
    }).sort_values('ะะฐะถะฝะพััั', ascending=False)
    
    print(importance_df.to_string(index=False))
    
    plt.figure(figsize=(10, 6))
    plt.barh(importance_df['ะัะธะทะฝะฐะบ'][:10], importance_df['ะะฐะถะฝะพััั'][:10])
    plt.xlabel('ะะฐะถะฝะพััั ะฟัะธะทะฝะฐะบะฐ')
    plt.title(f'ะขะพะฟ-10 ะฒะฐะถะฝะตะนัะธั ะฟัะธะทะฝะฐะบะพะฒ (ะณะปัะฑะธะฝะฐ={best_depth})')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
    
    return best_tree, best_depth, best_score
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

def decision_tree_regression_study(X, y, dataset_name="ะะฐะฝะฝัะต"):
    print(f"\n" + "="*70)
    print(f"ะะะจะะฎะฉะะ ะะะะะะ ะะะะะะกะกะะฏ - {dataset_name}")
    print("="*70)
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42
    )
    
    print("๐ ะััะปะตะดะพะฒะฐะฝะธะต ะฒะปะธัะฝะธั ะณะปัะฑะธะฝั ะดะตัะตะฒะฐ ะฝะฐ Rยฒ score...")
    depth_range = range(1, 21)
    train_scores = []
    test_scores = []
    
    for depth in depth_range:
        tree = DecisionTreeRegressor(max_depth=depth, random_state=42)
        tree.fit(X_train, y_train)
        train_scores.append(tree.score(X_train, y_train))
        test_scores.append(tree.score(X_test, y_test))
    
    best_depth_index = np.argmax(test_scores)
    best_depth = depth_range[best_depth_index]
    best_score = test_scores[best_depth_index]
    best_train_score = train_scores[best_depth_index]
    
    print(f"๐ฏ ะะตะทัะปััะฐัั:")
    print(f"  ะะฟัะธะผะฐะปัะฝะฐั ะณะปัะฑะธะฝะฐ: {best_depth}")
    print(f"  ะัััะธะน Rยฒ score: {best_score:.4f}")
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(depth_range, train_scores, label='ะะฑััะฐััะฐั ะฒัะฑะพัะบะฐ', marker='o', linewidth=2)
    plt.plot(depth_range, test_scores, label='ะขะตััะพะฒะฐั ะฒัะฑะพัะบะฐ', marker='s', linewidth=2)
    plt.axvline(x=best_depth, color='red', linestyle='--', alpha=0.7, label=f'ะัััะฐั ะณะปัะฑะธะฝะฐ={best_depth}')
    plt.xlabel('ะะฐะบัะธะผะฐะปัะฝะฐั ะณะปัะฑะธะฝะฐ ะดะตัะตะฒะฐ')
    plt.ylabel('Rยฒ Score')
    plt.title(f'ะะตัะฐััะตะต ะดะตัะตะฒะพ - ะะปะธัะฝะธะต ะณะปัะฑะธะฝั\n{dataset_name}', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    best_tree = DecisionTreeRegressor(max_depth=best_depth, random_state=42)
    best_tree.fit(X_train, y_train)
    y_pred = best_tree.predict(X_test)
    
    plt.subplot(1, 2, 2)
    plt.scatter(y_test, y_pred, alpha=0.6, s=50)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
    plt.xlabel('ะคะฐะบัะธัะตัะบะธะต ะทะฝะฐัะตะฝะธั')
    plt.ylabel('ะัะตะดัะบะฐะทะฐะฝะฝัะต ะทะฝะฐัะตะฝะธั')
    plt.title(f'ะัะตะดัะบะฐะทะฐะฝะธั vs ะคะฐะบัะธัะตัะบะธะต ะทะฝะฐัะตะฝะธั\nRยฒ = {best_score:.4f}', fontsize=12)
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n๐ ะะะะะะ ะะะะะะ ะะะจะะฎะฉะะะ ะะะะะะ:")
    print(f"  ะะฟัะธะผะฐะปัะฝะฐั ะณะปัะฑะธะฝะฐ ะดะตัะตะฒะฐ: {best_depth}")
    print(f"  Rยฒ score ะฝะฐ ัะตััะพะฒะพะน ะฒัะฑะพัะบะต: {best_score:.4f}")
    print(f"  Rยฒ score ะฝะฐ ะพะฑััะฐััะตะน ะฒัะฑะพัะบะต: {best_train_score:.4f}")
    
    overfitting_gap = best_train_score - best_score
    print(f"  ะะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test Rยฒ: {abs(overfitting_gap):.4f}")
    
    if overfitting_gap > 0.1:
        print("  โ๏ธ ะะพะทะผะพะถะฝะพะต ะฟะตัะตะพะฑััะตะฝะธะต (ะฑะพะปััะฐั ัะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test)")
    else:
        print("  โ ะะพะดะตะปั ัะพัะพัะพ ะพะฑะพะฑัะฐะตั (ะผะฐะปะฐั ัะฐะทะฝะธัะฐ ะผะตะถะดั train ะธ test)")
    
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    
    print(f"\n๐ ะะะะะะะะขะะะฌะะซะ ะะะขะะะะ ะะะะะะกะกะะ:")
    print(f"  MSE (Mean Squared Error): {mse:.4f}")
    print(f"  RMSE (Root Mean Squared Error): {rmse:.4f}")
    print(f"  MAE (Mean Absolute Error): {mae:.4f}")
    
    if best_score >= 0.7:
        print("  ๐ ะัะปะธัะฝะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (Rยฒ โฅ 0.7)")
    elif best_score >= 0.5:
        print("  ๐ ะฅะพัะพัะตะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.5 โค Rยฒ < 0.7)")
    elif best_score >= 0.3:
        print("  ๐ ะฃะดะพะฒะปะตัะฒะพัะธัะตะปัะฝะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (0.3 โค Rยฒ < 0.5)")
    else:
        print("  ๐ก ะะธะทะบะพะต ะบะฐัะตััะฒะพ ะผะพะดะตะปะธ (Rยฒ < 0.3)")
    
    print(f"\n๐ณ ะะะะะะกะขะฌ ะะะะะะะะะ:")
    feature_importances = best_tree.feature_importances_
    feature_names = X.columns if hasattr(X, 'columns') else [f'ะัะธะทะฝะฐะบ {i}' for i in range(X.shape[1])]
    
    importance_df = pd.DataFrame({
        'ะัะธะทะฝะฐะบ': feature_names,
        'ะะฐะถะฝะพััั': feature_importances
    }).sort_values('ะะฐะถะฝะพััั', ascending=False)
    
    print(importance_df.to_string(index=False))
    
    plt.figure(figsize=(10, 6))
    plt.barh(importance_df['ะัะธะทะฝะฐะบ'][:10], importance_df['ะะฐะถะฝะพััั'][:10])
    plt.xlabel('ะะฐะถะฝะพััั ะฟัะธะทะฝะฐะบะฐ')
    plt.title(f'ะขะพะฟ-10 ะฒะฐะถะฝะตะนัะธั ะฟัะธะทะฝะฐะบะพะฒ (ะณะปัะฑะธะฝะฐ={best_depth})')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
    
    print(f"\n๐ ะะะะะะ ะะจะะะะ:")
    residuals = y_test - y_pred
    
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 3, 1)
    plt.scatter(y_pred, residuals, alpha=0.6)
    plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    plt.xlabel('ะัะตะดัะบะฐะทะฐะฝะฝัะต ะทะฝะฐัะตะฝะธั')
    plt.ylabel('ะััะฐัะบะธ (ะคะฐะบั - ะัะพะณะฝะพะท)')
    plt.title('ะััะฐัะบะธ vs ะัะตะดัะบะฐะทะฐะฝะธั')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 2)
    plt.hist(residuals, bins=20, edgecolor='black', alpha=0.7)
    plt.xlabel('ะััะฐัะบะธ')
    plt.ylabel('ะงะฐััะพัะฐ')
    plt.title('ะะฐัะฟัะตะดะตะปะตะฝะธะต ะพััะฐัะบะพะฒ')
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 3, 3)
    plt.scatter(range(len(residuals)), residuals, alpha=0.6)
    plt.axhline(y=0, color='r', linestyle='--', alpha=0.5)
    plt.xlabel('ะะพะผะตั ะฝะฐะฑะปัะดะตะฝะธั')
    plt.ylabel('ะััะฐัะบะธ')
    plt.title('ะััะฐัะบะธ ะฟะพ ะฟะพััะดะบั')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print(f"  ะกัะตะดะฝะธะน ะพััะฐัะพะบ: {np.mean(residuals):.4f}")
    print(f"  ะกัะฐะฝะดะฐััะฝะพะต ะพัะบะปะพะฝะตะฝะธะต ะพััะฐัะบะพะฒ: {np.std(residuals):.4f}")
    
    return best_tree, best_depth, best_score
```
